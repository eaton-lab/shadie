### Element types

```python
import shadie
```

### Mutations

```python
m0 = shadie.mtype(0.5, 'n', 4, 1)
m0.draw();
```

```python
# create list of all mutations in a simulation
mlist = shadie.mlist(
    shadie.mtype(0.5, 'n', 2, 2),
    shadie.mtype(0.5, 'n', 4, 1),
    shadie.mtype(0.1, 'g', -3.0, 2),
    shadie.mtype(0.5, 'e', 2.5),
    shadie.mtype(0.5, 'f', 0.1),
)
```

```python
mlist.draw(width=400);
```

### Elements
Applies weighting to multiple mutation types.

```python
e0 = shadie.etype(
    mutations=[mlist[0], mlist[1]],
    frequencies=[5, 1],
)
e0.draw();
```

```python
e1 = shadie.etype(
    mutations=[mlist[0], mlist[1]],
    frequencies=[1, 5]
)
e1.draw();
```
