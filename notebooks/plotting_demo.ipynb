```python
import toyplot as tp
import pandas as pd
import numpy as np
```

## Generate Random Gene Info to Plot

```python
def random_gene():
    "generates a random gene with between 2 and 4 exons"
    
    noncoding = np.random.randint(2,100,1)
    exon1 = np.random.randint(2,80,1)
    intron1 = np.random.randint(2,60,1) 
    exon2 = np.random.randint(2,100,1)
    intron2 = np.random.randint(0,40,1)
    if intron2 != 0:
        exon3=np.random.randint(0,100,1)
    else:
        exon3=0
    if exon3 !=0:
        intron3=np.random.randint(0,50,1)
    else:
        intron3=0
    if intron3!=0:
        exon4=np.random.randint(2,50,1)
    else:
        exon4=0
    noncoding2=np.random.randint(2,100,1)
    
    return [noncoding, exon1, intron1, exon2, intron2, exon3, intron3, exon4, noncoding2]

```

## Make plots of single genes

```python
dna = np.column_stack((random_gene()))
canvas = tp.Canvas(width=900, height=150)
color = ["grey", "blue", "red", "blue", "red", "blue", "red", "blue", "grey"]
axes = canvas.cartesian(xlabel="number of basepairs")
mark = axes.bars(dna, along='y', color=color, baseline='stacked')
axes.y.show=False

dna2 = np.column_stack((random_gene()))
canvas = tp.Canvas(width=900, height=150)
color = ["grey", "blue", "red", "blue", "red", "blue", "red", "blue", "grey"]
axes = canvas.cartesian(xlabel="number of basepairs")
mark = axes.bars(dna2, along='y', color=color, baseline='stacked')
axes.y.show=False
    
```
